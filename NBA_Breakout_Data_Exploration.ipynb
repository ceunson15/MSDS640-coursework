```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('/content/NBA_Breakout.csv')
```

```python
df.head(10)
```

## Filling in Missing Values

```python
df.isnull().sum()
```

```python
df2 = df.drop(columns=['Awards'])
```

```python
df3 = df2.fillna(0)
```

```python
df3.isnull().sum()
```

## Summary Statistics

```python
df3.info()
```

```python
df3.describe().T
```

```python
df3.dtypes
```

## Plotting Distributions

```python
# Histograms of all features
df3.hist(figsize=(14, 10), bins=20, edgecolor='black', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df3.plot(x="Age", y="PTS", kind="scatter", alpha=0.6)
plt.xlabel("Age of NBA Players")
plt.ylabel("Points Scored Per Game (PTS)")
plt.title("Age vs Points")
plt.show()
```

```python
df3.plot(x="MPAVG", y="GS", kind="scatter", alpha=0.6)
plt.xlabel("Average Minutes Played Per Game")
plt.ylabel("Number of Games as a Starter")
plt.title("Avg Minutes vs Starting Player")
plt.show()
```

```python
df3.plot(x="eFG%", y="MPTotal", kind="scatter", alpha=0.6)
plt.xlabel("Effective Field Goal Percentage")
plt.ylabel("Total Minutes Played in a Season")
plt.title("Effective FG% vs Total Playtime")
plt.show()
```
